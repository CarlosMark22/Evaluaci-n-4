<h1> Análisis del problema propuesto </h1>

Este problema se trata de un movimiento oscilatorio. Antes de aplicar una fuerza F(t) en la masa $m_1$, el sistema se encuentra en reposo porque los resortes se encuentran en su longitud natural. Luego de aplicarse la fuerza F(t), el resorte con constante elástica $k_1$ se estira, provocando una fuerza elástica en oposición a la fuerza F(t). El amortiguador $b_1$ también generaría una fuerza en oposición. El resorte $k_2$ se comprimiría en el instante inicial debido al movimiento de $m_1$, lo que genera una fuerza en oposición en la masa $m_1$ y una fuerza en sentido positivo con respecto a $x_2$ y sobre la masa $m_2$. Ésta fuerza sobre $m_2$ es la que genera movimiento en la masa $m_2$. Luego de que la masa $m_2$ comienza a moverse, el resorte comienza a estirarse. Este proceso es repetitivo, lo que provoca que se trate de un movimiento vibratorio.

<h1> Explicación de diferencias de métodos </h1>

Podríamos iniciar diciendo que en el método de Lagrange se tienen los mismos principios de Isaac Newton, pero con una distinta formulación, siendo ésta mucho más sofisticada aplicando el método de Lagrange. Podemos nombrar algunos aspectos en los que el método de Lagrange resulta ser más efectivo: simplemente se involucra a las fuerzas que dan lugar al movimiento del sistema mecánico (Fuerzas externas y no conservativas); definimos una función escalar conocida como Lagrangiana (L = T – V; T: Energía Cinética, V: Energía Potencial), de la que se obtienen las ecuaciones del movimiento al derivar la función escalar, además estas ecuaciones son tantas como el número de coordenadas generalizadas; el número de coordenadas generalizadas se lo puede tener simplemente con un pequeño pero vital análisis de los grados de libertad del sistema, dado que el número de grados de libertad del sistema es igual al número de coordenadas generalizadas.

Por otra parte, el método de resolución por segunda Ley de Newton resulta más corto, sin embargo, es muy importante conocer cuál es el sentido del movimiento y de las fuerzas en los resortes. Al tratarse de un movimiento oscilatorio o vibratorio, es importante realizar una suposición inicial en la dirección de los resortes, pues el resorte $k_2$ podría estar comprimido o estirado. En este sentido resulta complicado estimar con exactitud la dirección de la fuerza elástica provocada por el resorte $k_2$ en un instante de tiempo y únicamente aplicando la Segunda Ley de Newton. Sin embargo, realizando las suposiciones adecuadas y colocando correctamente el sentido de las fuerzas, se puede obtener las ecuaciones del movimiento de manera rápida.

<h1> Conclusión del mejor método </h1>
Si bien el método de Lagrange resulta ser sencillo pero relativamente extenso, consideramos que es el mejor método para la resolución de este ejercicio, puesto que se evitan análisis como la construcción del DCL, colocar una adecuada dirección a todas las fuerzas que actúan en el sistema, realizar suposiciones del movimiento, etc. Es cierto que la resolución con Leyes de Newton resulta más rápida y sencilla, sin embargo, muchos se equivocarían al colocar la dirección correcta de las fuerzas elásticas.
