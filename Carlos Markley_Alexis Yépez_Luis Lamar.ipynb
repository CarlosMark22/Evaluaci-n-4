<h1> Explicacin de diferencias de mtodos </h1>

Podramos iniciar diciendo que en el mtodo de Lagrange se tienen los mismos principios de Isaac Newton, pero con una distinta formulacin, siendo sta mucho ms sofisticada aplicando el mtodo de Lagrange. Podemos nombrar algunos aspectos en los que el mtodo de Lagrange resulta ser ms efectivo: simplemente se involucra a las fuerzas que dan lugar al movimiento del sistema mecnico (Fuerzas externas y no conservativas); definimos una funcin escalar conocida como Lagrangiana (L = T  V; T: Energa Cintica, V: Energa Potencial), de la que se obtienen las ecuaciones del movimiento al derivar la funcin escalar, adems estas ecuaciones son tantas como el nmero de coordenadas generalizadas; el nmero de coordenadas generalizadas se lo puede tener simplemente con un pequeo pero vital anlisis de los grados de libertad del sistema, dado que el nmero de grados de libertad del sistema es igual al nmero de coordenadas generalizadas.

Por otra parte, el mtodo de resolucin por segunda Ley de Newton resulta ms corto, sin embargo, es muy importante conocer cul es el sentido del movimiento y de las fuerzas en los resortes. En este sentido resulta complicado estimar con exactitud la direccin de la fuerza elstica provocada por el resorte $k_2$ en un instante de tiempo y nicamente aplicando la Segunda Ley de Newton. Sin embargo, realizando las suposiciones adecuadas y colocando correctamente el sentido de las fuerzas, se puede obtener las ecuaciones del movimiento de manera rpida.
